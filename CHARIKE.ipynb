## Define perturbations class and title

```python
class perturbations:
    def __init__(self, values, units="", xlabel="", name="", stormname=""):
        self.values = values
        self.pointA = [None] * len(values)
        self.max_vol         = np.empty(len(values))
        self.max_vol_in_ctrl = np.empty(len(values))
        self.area            = np.empty(len(values))
        self.length_scale    = np.empty(len(values))
        self.units = units
        self.name = name
        if units != "": xlabel += ' ('+units+')' 
        self.xlabel = xlabel
        self.stormname = stormname
        self.index = -1
        self.len = len(values)
        # default directory name list
        self.dirname = ['veer'+p for p in values]
        if name == 'speeds': self.dirname = ['speed'+p for p in values]
        if name == 'vmaxes': self.dirname = ['vmax_PcAdjust'+p for p in values]
        if name == 'rmaxes': self.dirname = ['rmax'+p for p in values]
        
    def __iter__(self):
        self.index = -1
        return self
    def next(self):
        if self.index == len(self.values)-1:
            raise StopIteration
        self.index += 1
        return self.values[self.index]


stormname = "CHARIKE"
# Define domain (must have been created by NCL scripts already)
if stormname == 'CHARIKE': domain = 'stride02.-99.0E-87.0E25.0N31.0N'
if stormname == 'IKE': domain = 'stride10.-99.0E-87.0E25.0N31.0N'
if stormname == 'CHARLEY': domain = 'stride02.-85.0E-79.0E24.0N30.0N'
# Change to the directory containing all the perturbation subdirectories.
os.chdir('/glade/p/work/ahijevyc/ADCIRC/'+stormname+'/')

# Get title from attributes of length_scale variable in control run.
ifile='control/control.notideFalse_1.00m.'+domain+'.timeseries.nc'
fh = Dataset(ifile, mode='r')
ls = fh.variables['length_scale']
title =stormname+' inundation volume\ninundation depth threshold: '+str(ls.depth_threshold)+ls.depth_threshold_units+'\nlength scale left side: '+str(ls.left_percentile)+'%; right: '+str(ls.right_percentile)+'%'
fh.close()

```

## Time Series of Inundation Volume

```python
import re

# Define speed perturbations
# find all the directories, scrape off numeric part, sort numerically
values = [i[5:] for i in glob.glob('speed[+-]*[0-9]')]
speeds = perturbations(sorted(values, key=lambda item: float(item)), units='%', xlabel="along-track speed change",
                       name='speeds')

# Define veer perturbations
deg_symbol = u'\N{DEGREE SIGN}'
values = [i[4:] for i in glob.glob('veer[+-]*[0-9]')]
veers = perturbations(sorted(values, key=lambda item: float(item)), 
                      units=deg_symbol+'CW day$^{\mathrm{-1}}$', stormname=stormname,
                      name='veers',xlabel='cross-track error rate')

# Define vmax perturbations
values = [i[13:] for i in glob.glob('vmax_PcAdjust[+-][0-9]')]
vmaxes = perturbations(sorted(values, key=lambda item: int(item)), name='vmaxes', 
                       units = "(sort-of) kts hr$^{\mathrm{-1}}$", xlabel = "vmax change")

# Define rmax perturbations
values = [i[4:] for i in glob.glob('rmax[+-]*[0-9]')]
rmaxes = perturbations(sorted(values, key=lambda item: int(item)), units = "%", xlabel="rmax change",
                       name='rmaxes')

A = [0, 64738 ,112521, 118019 ]
if stormname == 'IKE' or stormname == 'CHARIKE': # A is a list of nodes that experience large inundation.
    A = [82267,81635,384783,152844,152987,374075,355075,285018,285019,152899,414206,348971]
    A = [81635, 152987,374075]
    A = [int(i/2) for i in [109073,152899, 440403]]

    

# Get lat/lon of nodes in list A.
#tried to get lat and lon of points A but point As were drawn from 
#perfectmodel, which doesn't necessary have the same domain/stride.
ifile='control/fort.63.'+domain+'.nc'
if stormname == 'IKE': ifile = re.sub(r"stride\d\d","stride02",ifile)
print "opening ", os.path.dirname(os.path.realpath(ifile)), " to find A points ", A
fh = Dataset(ifile, mode='r')
print ifile, fh
lon = fh.variables['x'][A]
lat = fh.variables['y'][A]
fh.close()
#print lon, lat

for perturbation in [veers, vmaxes, rmaxes, speeds]:
    fig, ax = plt.subplots()
    plt.title(title)
    plt.tight_layout(rect=(0.04,0.09,0.93,0.96))
    print perturbation.values, perturbation.name, perturbation.dirname
    vdirs = perturbation.dirname
    for i, v in enumerate(perturbation):
        files = glob.glob(vdirs[i]+'/*.notideFalse_1.00m.'+domain+'.timeseries.nc')
        if len(files)!=1: print "should have 1 file in "+vdirs[i]
        for nc_file in files:
            fh = Dataset(nc_file, mode='r')
            inund = fh.variables['inundation_volume']
            imax = np.argmax(inund[:])
            perturbation.max_vol[i] = inund[imax]
            perturbation.max_vol_in_ctrl[i] = fh.variables['volume_in_ctrl'][imax]
            perturbation.length_scale[i] = fh.variables['length_scale'][imax]
            perturbation.area[i] = fh.variables['inundation_area'][imax]
            time = fh.variables['time']
            base_date = dt.datetime.strptime(time.base_date, "%Y-%m-%d %H:%M:%S %Z")
            dt_time = [base_date+dt.timedelta(0,t) for t in time]
            thick = 3 if float(v) == 0 else 1
            plt.plot(dt_time,inund[:],label=v,linewidth=thick,marker='^'if float(v)<0 else '',markersize=3)
            units = getattr(inund,'units')
            fh.close()
    if units=="km~S~3~N~": units = '$\mathregular{km^3}$'
    plt.ylabel(units)

    plt.grid()
    ax.xaxis.set_major_formatter(dates.DateFormatter('%-m/%-d %-H'+'Z'))
    ax.xaxis.set_major_locator(dates.HourLocator(byhour=[0,6,12,18]))
    fig.autofmt_xdate()
    plt.xlabel(base_date.strftime("%Y"))
    
    # shrink current axis by 15%
    box = ax.get_position()
    ax.set_position([box.x0,box.y0,box.width*0.85,box.height])
    
    # put legend to the right of the current axis
    plt.legend(loc='center left', bbox_to_anchor=(1,0.5), title=perturbation.name+"\n"+perturbation.units)
    status = mysavfig(perturbation.name+'_timeseries.png',dpi=225)
```

### Pearson correlation coefficient squared, Bias, Root Mean Squared Error

```python
for perturbation in [veers, vmaxes, rmaxes, speeds]:
    r2 = []
    bias = []
    rmse = []
    fig, ax = plt.subplots()
    plt.tight_layout(rect=(0.02,0.05,0.89,0.9))

    pdirs = perturbation.dirname
    for i, p in enumerate(perturbation):
        pdir = pdirs[i]
        for nc_file in glob.glob(pdir+'/*.perfectmodel.zoom1.nc'):
            print 'reading '+nc_file
            fh = Dataset(nc_file, mode='r')
            r2.append(fh.variables['rc_tmp'].getncattr('r2'))
            yave = fh.variables['rc'].getncattr('yave')
            xave = fh.variables['rc'].getncattr('xave')
            bias.append(yave/xave)
            obs = fh.variables['obs'][:] # tried np.asarray() but it removed _FillValue
            model = fh.variables['model'][:]
            rmse.append(np.sqrt(np.mean((obs - model)**2)))
            print model.shape
            As = np.argmax(model[model<6.5]) # used to find nodes to put in list A (above)
            print As, model[As], len(perturbation.pointA), perturbation.index, i, model[A]
            perturbation.pointA[i] = model[A]
            fh.close()

    print perturbation.values, r2
    plt.plot(perturbation.values,r2,label='$\mathregular{r^2}$',marker='o')
    plt.title(title)
    plt.plot(perturbation.values,bias,label='multiplic. bias',marker='o')
    plt.title(title)
    plt.plot(perturbation.values,rmse,label='rmse [m]',marker='o')
    plt.title(title)
    ax.yaxis.grid()
    plt.xlabel(perturbation.xlabel)
    plt.ylabel('correlation coeff., bias, or meters')

    # shrink current axis by 15%
    box = ax.get_position()
    ax.set_position([box.x0,box.y0,box.width*0.85,box.height])
    
    # put legend to the right of the current axis
    plt.legend(loc='center left', bbox_to_anchor=(1,0.5),fontsize=10)

    status = mysavfig(perturbation.name+'_stats.png',dpi=225)
```
